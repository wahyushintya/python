```python
#Soal nomor  1

#Menghitung Luas Persegi
S = int(input("Masukan Sisi Persegi : "))
Luas_P = S*S

#Menghitung Luas Lingkaran
phi = 3.14
R = int(input("Masukan panjang jari jari lingkaran : "))
Luas_L = phi*R*R

# luas daerah yang diarsir
L_Diarsir = Luas_P - Luas_L

print("luas daerah yang diarsir adalah = ", L_Diarsir)
```
