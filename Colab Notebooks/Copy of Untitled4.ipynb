```python
print("Menampilkan program ganjil genap")
print("Tampilkan bilangan\n""1. Ganjil\n""2. Genap")

pilih=int(input("masukan pilihan (1/2): "))

if pilih == 1:
  print("bilangan ganjil")
  awal=int(input("Masukan Nilai awal = "))
  akhir=int(input("Masukan nilai akhir = "))
  for i in range(awal,akhir):
    if(i % 2 ==1):
      print(i, end=" ")

elif pilih == 2:
  print("bilangan genap")
  awal=int(input("Masukan Nilai awal = "))
  akhir=int(input("Masukan nilai akhir = "))
  for i in range(awal,akhir):
    if(i % 2 ==0):
      print(i, end=" ")

else:
  print("Tidak valid")
```
