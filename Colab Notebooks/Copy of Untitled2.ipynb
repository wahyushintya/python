```python
print("=======MENGHITUNG LUAS BANGUNAN=======")
print("1. Persegi \n2. Persegi Panjang \n3. Segitiga \n4. Lingkaran")
pilih = input("ingin menghitung yg mana? :")

if pilih == "1":
  print("ANDA MEMILIH BANGUNAN PERSEGI")
  sisi= int(input("Masukan sisi"))
  Luas= sisi*2
  print("jadi luasnya adalah ", Luas)

elif pilih == "2":
  print("ANDA MEMILIH BANGUNAN PERSEGI PANJANG")
  p = int(input("Masukan panjang :"))
  l = int(input("Masukan Lebar :"))
  Luas = p*l
  print("jadi luasnya adalah ", Luas)

elif pilih == "3":
  print("ANDA MEMILIH BANGUNAN SEGITIGA")
  a = int(input("Masukan alas :"))
  t = int(input("Masukan tinggi :"))
  Luas = 1/2 *a *t
  print("jadi luasnya adalah ", Luas)

elif pilih == "4":
  print("ANDA MEMILIH BANGUNAN LINGKARAN")
  r = int(input("Masukan ruas :"))
  Luas = 3.14 * r*2
  print("jadi luasnya adalah ", Luas)

else:
   print("APAAA?")
```
