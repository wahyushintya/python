```python
angka = [30,90,84,2,10,44,77,123]
print(angka)

#menambahkan angkat 100 dan 200 di akhir list
angka.append(100)
angka.append(200)
print(angka)

#menghapus angka pada index 5
angka.pop(5)
print(angka)

#tambahkan angka kemballi angka 1 dinindex pertama
angka.insert(0, 1)
print(angka)

#hitung panjang list
print("panjang deret =", len(angka))

#menamilkan nilai terbesar dari seluruh nilai
print(max(angka))

#hitung total keseluruhan
print(sum(angka))
```

```python
#list 2 dimensi

list_siswa = [
    ["agus","budi","mia", "dewa"],
    ["putri","wayan","fahri","made"],
    ["ana","mona","adele", "candra"]]

for baris in list_siswa:
  for kolom in baris:
    print(kolom, end=" ")
  print()
print("")

print("Menampilkan nama fahri saja : ", list_siswa[1][2])

print("Menampilkan nama Dewa : ", list_siswa[0][3])
print("Menampilkan nama Made: ",list_siswa[1][3])
print("Menampilkan nama adele: ",list_siswa[2][2])
```

```python
#list 2 dimensi

list_siswa = [
    ["agus","budi","mia", "dewa"],
    ["putri","wayan","fahri","made"],
    ["ana","mona","adele", "candra"]]
for row in list_siswa :
  print(row)
print("")

print("Menampilkan nama fahri saja : ", list_siswa[1][2])

print("Menampilkan nama : ", list_siswa[0][3], list_siswa[1][3], list_siswa[2][2])
```

```python
matriks1 = [
    [5, 8],
    [3, 9]
]

matriks2 = [
    [2, 1],
    [4, 2]
]

# Inisialisasi matriks hasil dengan elemen nol
hasil = [[0, 0], [0, 0]]

# Menjumlahkan elemen-elemen matriks
for i in range(len(matriks1)):
    for j in range(len(matriks1[0])):
        hasil[i][j] = matriks1[i][j] + matriks2[i][j]

# Menampilkan hasil penambahan
print("Hasil Penambahan Matriks:")
for baris in hasil:
    print(baris)
```
