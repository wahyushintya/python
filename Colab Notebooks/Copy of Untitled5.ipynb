```python
def cek_tahun_kabisat(tahun):
    if (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0):
        return True
    else:
        return False

while True:
    try:
        input_tahun = int(input("Masukkan tahun (0 untuk keluar): "))

        if input_tahun == 0:
            print("Keluar dari program.")
            break

        if cek_tahun_kabisat(input_tahun):
            print(f"{input_tahun} adalah tahun kabisat.")
        else:
            print(f"{input_tahun} bukan tahun kabisat.")

    except ValueError:
        print("Masukkan harus berupa angka. Coba lagi.")

```
