```python
nilai = int(input("Masukan jumlah data = "))
i = 0
data = []
for o in range (i, nilai) :
  i += 1
  a = int(input(f"Masukan data ke-{i}: "))
  data.append(a)
print(data)

jumlah = 0
for a in data :
  jumlah += a
print("Jumlah seluruh data = ", jumlah)

ratarata = jumlah/nilai
print("Rata ratanya adalah = ", ratarata)
```
