```python
#Soal 1
#menghitung luas dan keliling persegi
s= 10
L_P = s*s
K_P = 2*s

#luas dan keliling setengah lingkaran
r = 5
L_setling = 1/2 * 3.14 *r *r
K_Setling = 1/2 * 2 * 3.14 * r

#total
L_total = L_P + L_setling + L_setling
K_total = K_P + K_Setling + K_Setling

#luas bangun
print("Luas bangun biru tersebut adalah ", L_total, "Cm^2")
print("Keliling bangun biru tersebut adalah ", round(K_total,1), "Cm")

```

```python
#Soal 3
#Menghitung luas lingkaran
import math

diameter = 12
jari_jari = diameter / 2
Luas_Ling = 3.14 * (jari_jari**2)
Luas_Bulat = round(Luas_Ling)

print("Luas Lingkaran tersebut : ", Luas_Ling,"Cm^2", "dan dibulatkan menjadi ", Luas_Bulat, "cm^2")
```

```python
#Soal 4

#Menentukan hadiah dari total belanja
Belanja = int(input("Masukan Total Belanjaan Anda : "))

if Belanja < 500000 :
  print("Selamat anda mendapatkan hadiah kaos kaki")
elif Belanja == 500000 :
  print("Selamat anda mendapatkan hadiah alat tulis")
elif Belanja > 500000 and Belanja <= 1000000 :
  print("Selamat anda mendapatkan hadiah baju kaos")
elif Belanja > 1000000 :
  print("Selamat anda mendapatkan hadiah sepeda")
else :
  print(" Anda tidak mendapat diskon")
```
